# Step 2 of MSM construction: dimensionality reduction

tICA vs PCA


Original feature sets dimensions:
- dihedrals: 66
- contacts: 42


Compression goal:
- 2d for visualization
- ~ 5d for MSM construction

Low dimensionality facilitates clustering, which is step 3 of MSM construction.

```python
data_folder = "./data/"

import numpy as np
import deeptime as deep
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import matplotlib.cm as cm
from matplotlib.colors import ListedColormap
binary_cmap = ListedColormap(['white', 'black'])

data_folder = "./data/"


#################################################
#################################################
#################################################

def free_energy_1d(y, bins = 100):
    hist, edges = np.histogram(y, bins = bins, density=True)
    centers = 0.5 * (edges[:-1] + edges[1:])
    p = hist + 1e-12
    F = - np.log(p)
    F -= F.min()
    return centers, F
```

### Load feature sets

```python
# set 1: distances
distances = np.loadtxt(data_folder + "hp35.mindists2", delimiter = " ", dtype = float)
print("Data shape", distances.shape) # [nm] nanometers
print("Data size (MB)", distances.nbytes / (10**6))

# set 2: dihedrals
#dihedrals = np.loadtxt(data_folder + "hp35.dihs.shifted", delimiter = " ", dtype = float)
#print(dihedrals.shape) # [radians] 
```

### MoSAIC clusters.

```python
contacts = [
    tuple(x)
    for x in np.loadtxt(
        data_folder + "hp35.mindists2.ndx",
        comments="#",
        usecols=(0, 1),
        dtype=int
    )
]

cluster_dict = {
    1 : [(3,14), (3, 13), (6,14), (5, 14)],
    2: [(7, 12), (7, 13), (6,12), (7,11), (6, 11), (6, 17)],
    3: [(12,17), (12,16), (12,20), (13, 17)],
    4: [(18,25), (17,25), (20, 25)],
    5: [(24, 28), (20, 28), (25, 29)],
    6: [(29,35), (29,34), (30,35), (29,33)],
    7: [(10,34), (9,32), (10,29)]
}

def cluster_idx(contact_tuple):
    idx = None
    for i in cluster_dict.keys():
        if contact_tuple in cluster_dict[i]:
            idx = i
            continue
    if idx == None:
        idx = 8
    return idx


feature_index_list = [
    (
    contacts[i],
    cluster_idx(contacts[i])
    )
    for i in range(len(contacts))
]
distances[0]
```

## Sort features by cluster

```python
# permutation that sorts by cluster index
sort_idx = np.argsort(
    [x[1] for x in feature_index_list]
)

distances = distances[:, sort_idx]
contacts = [contacts[i] for i in sort_idx]
feature_index_list = [feature_index_list[i] for i in sort_idx]
distances[0]
```

## Preprocessing: smoothing the time series


Authors apply a Gaussian low-pass filter, which is basically a moving average with gaussian weights instead of equal weights. There are several theoretical reasons why a Gaussian filter should be preferrable to a simple MA.


However, the goal simply remains smoothing out the signal.


Why would you want to do that? See handwritten notes. It is expected to lower the uncertainties of cluster assignation and it is suggested by the authors as an alternative procedure to *dynamic core sets*.

```python
from scipy.ndimage import gaussian_filter1d

smoothed_distances = gaussian_filter1d(distances, sigma=2)


fig, ax = plt.subplots(figsize = (10, 3))
ax.plot(np.arange(distances.shape[0]), distances[:, 0], label = "raw")
ax.plot(np.arange(distances.shape[0]), smoothed_distances[:, 0], label = "smoothed")
ax.legend()
ax.set_xlabel("simulation frame idx")
ax.set_ylabel("distance [nm]")
```

## PCA

```python
X = smoothed_distances

pca = PCA()
pca.fit(X)
```

```python
eigvecs = pca.components_ # components, space dimension (42, 42), ordered by decreasing eigval
eigvals = pca.explained_variance_
```

```python
Y = pca.fit_transform(X) # rotate the dataset onto the orthonormal basis of eigenvectors
explained = pca.explained_variance_ratio_
cum_explained = np.cumsum(explained)
max_pcs = eigvecs.shape[0]         # shape: (n_pcs, n_features)
```

```python
fig, ax = plt.subplots(figsize = (5, 3))
ax.plot(np.arange(1, max_pcs + 1), cum_explained, marker = "o")
ax.set_xscale("log", base = 10)
ax.set_xlabel("number of PCs")
ax.set_ylabel("cumulative explained variance")
ax.grid(which = "both")
```

**Comment** I get different results from those of the authors: different eigenvectors and explained variance.

However, I still get the bimodal shape in the free energy curves of the first PCs, and the eigenvectors appear to describe quite interpretable motions.

- PC 1: transition folded<->unfolded: clearly this is the collective motion where the shape of the molecule changes the most, so it is expected to correspond to the first principal component, which indeed captures the motion of *highest amplitude*.
- PC 2: motion of the two terminals (highest eigvec components for RED, BLUE, YELLOW and CYAN clusters). This is indeed the highest amplitude equilibrium motion of the folded protein, as it appears from a visual inspection of the trajectory movie.
- PC 3: motion of the internal part (helices 2 and 3), clusters GREEN, ORANGE, PURPLE. Correspondingly, change in the helices-stabilizing contacts of cluster 8.
- PCs 6,7,8 all show significant amplitude in the contacts of cluster 8, which are the helix stabilizing contacts, thus capturing the motion of extension and compression of the helices.

### Dimensionality reduction


How many PCs? The authors use the following criteria:
- **Nonquadratic appearance of the free energy curves along the PCs**, 
- decay times of their autocorrelation function,
- explained percentage of the total correlation.


So, basically we are restricting our focus on the folding<->unfolding process. We want to build a MSM that describes **this** process, and metastable states should reflect **intermediate** states of **this** process.

We are uninterested in motions that happen *within* the folded basin. We are looking for the **intermediate** states in the folded <-> unfolded transition.

The free energy curve is therefore the primary criterion for deciding which PCs to include and which to exclude.

```python
npcs = 5
# non trivial free energy curves, cumulative variance explained is > 97%


Y_reduced = Y[:, 0:npcs].copy()
Y_reduced.shape


#### Free some memory #####
del Y
del X
```

```python
from itertools import combinations
pairs = list(combinations(np.arange(npcs), 2))




cols = npcs-1
rows = npcs-1
fig, axs = plt.subplots(nrows = rows, ncols = cols, figsize = (cols*5, 4*rows))

for ax in axs.ravel():
    ax.set_axis_off()

for idx, (i, j) in enumerate(pairs):
    ax = axs[i, j-1]
    ax.set_axis_on()
    ax.scatter(Y_reduced[:, i], Y_reduced[:, j], s = 0.001, alpha = 0.1)
    ax.set_xlabel(f"PC{i}")
    ax.set_ylabel(f"PC{j}")
```

**Comment**: The first PCs show two well separated clusters, that's a good sign!

## Clustering into MSM microstates (first of the two clustering steps)


Choice of the clustering algorithm.

- kmeans: then how to choose the number of clusters, which must be given as input parameter? A.W. uses GMRQ (Generalized Matrix Rayleigh Quotient).
- authors of ref.1 and ref.2. suggest hierarchical density based clustering.



Let's simply do **one** MSM using kmeans and an arbitrary chosen number of clusters, say $k=200$. For the moment, lets simply go through the pipeline. 

--------

The GMRQ method is quite simple to implement (less simple to understand!!). Basically you need to compute various MSM for different choices of hyperparameters (e.g. number of clusters, clustering algorithm) and score them against each other. Therefore it is instructive to first go through the code needed to build ONE MSM.


For GMRQ, see this example from MSMBuilder:

GitHub: msmbuilder/msmbuilder2022/examples/GMRQ-Model-Selection.ipynb

and the paper: McGibbon, R. T. and V. S. Pande, Variational cross-validation of slow dynamical modes in molecular kinetics (2014).

--------

```python
from tqdm.notebook import tqdm  # progress bar (optional)
from deeptime.clustering import KMeans

n_clusters = 200



kmeans = KMeans(
    n_clusters=n_clusters,
    init_strategy='uniform',
    max_iter=100,
    fixed_seed=13,
    n_jobs=8,
    progress=tqdm
)

#n_clusters = 200

################### Fit kmeans on the 5 PCs ##########################
kmeans.fit(Y_reduced)
```

```python
###### Discretize the trajectories: from feature values to microstate idx #########
discretized_trajectories = kmeans.transform(Y_reduced)
model = kmeans.fetch_model() # fetches the latest model, containing all important information


####### Save to file #####################
np.savez("kmeans_results", traj = discretized_trajectories, centers = model.cluster_centers)
```
